```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Q1. What is the maximum number of matches played by an individual player in a season?

```python
max_match=iplDataSet['Mat'].max()
iplDataSet[['Player','Mat']][iplDataSet['Mat']==max_match]
```

# Q2. What is the highest average of a player in a season?

```python
max_avg=iplDataSet['Avg'].max()
iplDataSet[['Player','Avg']][iplDataSet['Avg']==max_avg]
```

# Q3. What is the average number of matches played by the players in a season?

```python
avg_no_of_match_played=iplDataSet['Mat'].mean()
avg_no_of_match_played
```

# Q11. Plot a histogram of number of matches played in a season by players.

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.hist(iplDataSet['Mat'],histtype='stepfilled',edgecolor='black',color='g',bins=12)
plt.xlabel('Mat')
plt.ylabel('no. of player')
plt.title("number of matches played in a season ")
plt.show()

```

# Q12. Plot the histogram of balls faced by players.

```python
df1=iplDataSet[['BF']]
plt.hist(df1,histtype='stepfilled',edgecolor='black',)
plt.xlabel('BF')
plt.ylabel('Frequency')
plt.title("balls faced by players")
plt.show()
```

# Q13. Top 20 players with most runs in a season.

```python
# d=sorted(iplDataSet['Runs'],reverse=True)
dframe=iplDataSet[['Player','Runs']]
dframe.sort_values('Runs',ascending=False).head(20)
```

# Q14. Bar chart for top 20 players with best strike rate in a season.

```python
df1=iplDataSet[['Player','SR']].sort_values('SR',ascending=False).head(20)

df1.plot.barh()
```

# Q15. Top 30 players with highest number of sixes.

```python
iplDataSet[['Player','6s']].sort_values('6s',ascending=False).head(30)
```

# Q16. Scatter plot of runs scored by a player v/s balls faced in a season.

```python
df1=iplDataSet[['Player','BF']]
plt.figure(figsize=(25,25))
plt.scatter(df1['BF'],df1['Player'])
plt.xlabel('ball faced by player',fontsize=25)
plt.ylabel('players',fontsize=25)
plt.title('Player vs balls faced in a season',fontsize=25)
# df1.plot.scatter('BF','Player',alpha=1)
```
